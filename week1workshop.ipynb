```python
import numpy as np
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])
add = x+y
print("add:",add)
sub = x-y
print("sub:",sub)
mul = x*y
print("mul:",mul)
square = x**2
print("square:",square)
dot_vw = np.dot(v, w)
print("Dot product of v and w:", dot_vw)
dot_xv = np.dot(x, v)
print("Dot product of x and v:\n", dot_xv)
dot_xy = np.dot(x, y)
print("Dot product of x and y:\n", dot_xy)
concat_xy_row = np.concatenate((x, y), axis=0)
print("x and y concatenated along rows:\n", concat_xy_row)
v_col = v.reshape(-1, 1)
w_col = w.reshape(-1, 1)
concat_vw_col = np.concatenate((v_col, w_col), axis=1)
print("v and w concatenated along columns:\n", concat_vw_col)

concat_xv_fixed = np.concatenate((x, v.reshape(1,2)), axis=0)
print("x and v concatenated after reshaping v:\n", concat_xv_fixed)



```

```python
import numpy as np

A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

A_inv = np.linalg.inv(A)


identity = np.dot(A, A_inv)

print("A * A^-1 =\n", identity)
AB = np.dot(A, B)
BA = np.dot(B, A)

print("AB =\n", AB)
print("BA =\n", BA)
print("AB == BA ?", np.array_equal(AB, BA))
AB_T = np.dot(A, B).T
BT_AT = np.dot(B.T, A.T)

print("(AB)^T =\n", AB_T)
print("B^T A^T =\n", BT_AT)
print("(AB)^T == B^T A^T ?", np.array_equal(AB_T, BT_AT))


```
